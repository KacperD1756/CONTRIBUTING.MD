```python
import psycopg2

# Dane połączeniowe do bazy danych
db_config = {
    'host': 'localhost', 
    'dbname': 'postgres',
    'user': 'postgres',
    'password': '1234'
}

# Połączenie z bazą danych
conn = psycopg2.connect(**db_config)
cursor = conn.cursor()

# Przykład dodawania wierszy do tabeli 'llmmodels'
try:
    # Dodaj pierwszy wiersz

    cursor.execute(''' 
                    
                    DROP TABLE IF EXISTS llmmodels CASCADE;
                    DROP TABLE IF EXISTS qa_8k CASCADE;
                    DROP TABLE IF EXISTS qa_8k_llmmodels CASCADE;
                    DROP TABLE IF EXISTS qa_cat CASCADE;
                    DROP TABLE IF EXISTS qa_cat_llmmodels CASCADE;
                    DROP TABLE IF EXISTS qa_dialog CASCADE;
                    DROP TABLE IF EXISTS qa_dialog_llmmodels CASCADE;
                    DROP TABLE IF EXISTS qa_code CASCADE;
                    DROP TABLE IF EXISTS qa_code_llmmodels CASCADE;
                    DROP TABLE IF EXISTS qa_sources CASCADE;  
                    DROP TABLE IF EXISTS qa_sources_llmmodels CASCADE;  
                   
                    -- Tworzenie tabeli 'llmmodels'
                    CREATE TABLE llmmodels (
                        id SERIAL PRIMARY KEY,
                        name VARCHAR(255) NOT NULL,
                        path VARCHAR(255) NOT NULL,
                        prompt_template TEXT,
                        maxtokens INT
                    );
                   
                    -- Tworzenie tabeli 'qa_8k'
                    CREATE TABLE qa_8k (
                        id SERIAL PRIMARY KEY,
                        question TEXT NOT NULL,
                        answer TEXT NOT NULL,
                        domian varchar(255)
                    );
                     

                
                    -- Tworzenie tabeli 'qa_8k_llmmodels' z kluczami obcymi odnoszącymi się do 'qa-8k' i 'llmmodels'
                    CREATE TABLE qa_8k_llmmodels (
                        id SERIAL PRIMARY KEY,
                        qa_8k_id INT NOT NULL,
                        llmmodels_id INT NOT NULL,
                        language VARCHAR(255) NOT NULL,
                        similarity FLOAT NOT NULL,
                        answerLLM TEXT,
                        FOREIGN KEY (qa_8k_id) REFERENCES qa_8k(id),
                        FOREIGN KEY (llmmodels_id) REFERENCES llmmodels(id)
                    );                   
                   
                    -- Tworzenie tabeli 'qa-cat'
                    CREATE TABLE qa_cat (
                        id SERIAL PRIMARY KEY,
                        category VARCHAR(255) NOT NULL,
                        question TEXT NOT NULL,
                        answer TEXT NOT NULL
                    );
                    -- Tworzenie tabeli 'qa_cat_llmmodels' z kluczami obcymi odnoszącymi się do 'qa_cat' i 'llmmodels'
                        CREATE TABLE qa_cat_llmmodels (
                        id SERIAL PRIMARY KEY,
                        qa_cat_id INT NOT NULL,
                        llmmodels_id INT NOT NULL,
                        language VARCHAR(255) NOT NULL,
                        answerLLM TEXT,
                        FOREIGN KEY (qa_cat_id) REFERENCES qa_cat(id),
                        FOREIGN KEY (llmmodels_id) REFERENCES llmmodels(id)
                    );     
                    --Tworznei tabli 'qa-dialog'
                       CREATE TABLE qa_dialog (
                            id SERIAL PRIMARY KEY,
                            q1 TEXT NOT NULL,
                            q2 TEXT NOT NULL,
                            q3 TEXT NOT NULL,
                            q4 TEXT NOT NULL,         
                            a1 TEXT NOT NULL,
                            a2 TEXT NOT NULL,
                            a3 TEXT NOT NULL,   
                            a4 TEXT NOT NULL
                        );      
                    --Tworzenie tabeli 'qa-dialog_llmmodels' z kluczami obcymi odnoszącymi się do 'qa_dialog' i 'llmmodels'
                        CREATE TABLE qa_dialog_llmmodels (
                        id SERIAL PRIMARY KEY,
                        qa_dialog_id INT NOT NULL,
                        llmmodels_id INT NOT NULL,
                        language VARCHAR(255) NOT NULL,
                        aLLM1 TEXT NOT NULL,
                        aLLM2 TEXT NOT NULL,
                        aLLM3 TEXT NOT NULL,
                        aLLM4 TEXT NOT NULL,
                        FOREIGN KEY (qa_dialog_id) REFERENCES qa_dialog(id),
                        FOREIGN KEY (llmmodels_id) REFERENCES llmmodels(id)    
                    );
                    --Tworzenie tabeli 'qa_code'
                        CREATE TABLE qa_code (
                            id SERIAL PRIMARY KEY,
                            question TEXT NOT NULL,
                            answer TEXT NOT NULL
                        );
                    --Tworzenie tabeli 'qa_code_llmmodels' z kluczami obcymi odnoszącymi się do 'qa_code' i 'llmmodels'
                        CREATE TABLE qa_code_llmmodels (
                        id SERIAL PRIMARY KEY,
                        qa_code_id INT NOT NULL,
                        llmmodels_id INT NOT NULL,
                        language VARCHAR(255) NOT NULL,
                        answerLLM TEXT NOT NULL,
                        FOREIGN KEY (qa_code_id) REFERENCES qa_code(id),
                        FOREIGN KEY (llmmodels_id) REFERENCES llmmodels(id)
                    );  
                    --Tworzenie ta beli 'qa-sources
                        CREATE TABLE qa_sources (  
                            id SERIAL PRIMARY KEY,
                            question TEXT NOT NULL,
                            answer TEXT NOT NULL
                        );
                    --Tworzenie tabeli 'qa-sources_llmmodels' z kluczami obcymi odnoszącymi się do 'qa_sources' i 'llmmodels'   
                        CREATE TABLE qa_sources_llmmodels (
                        id SERIAL PRIMARY KEY,
                        qa_sources_id INT NOT NULL,
                        llmmodels_id INT NOT NULL,  
                        language VARCHAR(255) NOT NULL,
                        answerLLM TEXT NOT NULL,
                        FOREIGN KEY (qa_sources_id) REFERENCES qa_sources(id),
                        FOREIGN KEY (llmmodels_id) REFERENCES llmmodels(id)
                    );  
                   '''


)
   
    conn.commit()

    print("Dodano wiersze do tabeli 'llmmodels'.")

except psycopg2.Error as e:
    print("Wystąpił błąd podczas dodawania wierszy:", e)
    conn.rollback()
finally:
    cursor.close()
    conn.close()




```
